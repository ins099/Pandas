# More functions with writing csv

columns = ['col1', 'col2'] Save specific coulmns
headers = False Removes header from file 

```python
# reading excel
ex= pd.read_excel('stock.xlsx',"Sheet1")
ex
```

```python
#converters

def eps_converter(cell):
    if cell == "not available":
        return None
    else:
        return cell

ex= pd.read_excel('stock.xlsx',"Sheet1", converters = {
    'eps':eps_converter,
})
ex
```

```python
#writing excel
ex.to_excel ('new.xlsx', sheet_name = "stocks", index = False, startrow = 2, startcol = 3)
```

```python
#using excel writer class to write both DF in single file but two sheets
with pd.ExcelWriter('WeatherStocks.xlsx') as writer:
    df.to_excel(writer, sheet_name = 'Weather')
    ex.to_excel(writer, sheet_name = 'Stocks')
```

# Handling Missing Data

```python
df3 = pd.read_csv('weather_Data.csv',parse_dates = ['day'])
df3

#as we see that the type of Day col is string, in order to change into date, use this
# parse_dates = ['day'] 
#as argument while loading csv
```

```python
type(df3.day[0])
```

```python
df3.set_index('day', inplace =True)
df3
```

#Fillna

```python
#1: Filling NaN as 0 or any thing in whole table
df3.fillna(0)
```

```python
#only respective column depending on data
df3.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event',
})
```

```python
""" but looking at this table make sense that temperature and windspeed to be as 0 recorded which creates no valid 
points when reading the data, so we use these step"""

#FORWARD FILL 
df3.fillna(method = 'ffill')
```

```python
#COPY THE NEXT VALUE OF EMPTY CELL
df3.fillna(method= 'bfill', axis = 'columns', limit = 1) #copy data colomn to col
#limit function copy only the amount you want 
```

```python
# Values are so far away from each other they dont make sense, so in order to make it work
# we use linear interpolate, we can have this work as per need or better guess for missing data
```

```python
df3.interpolate(method = 'time')
```
